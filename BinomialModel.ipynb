```python
import numpy as np

# Parameters
S0 = 100      # current stock price
K = 100       # strike price
u = 1.2       # up factor
d = 0.8       # down factor
r = 0.05      # risk free rate

# Stock prices after one step
S_up = S0 * u
S_down = S0 * d

# Risk-neutral probability
p = ((1 + r) - d) / (u - d)

# Payoffs for a European Call Option
C_up = max(S_up - K, 0)
C_down = max(S_down - K, 0)

# Present value of option
C0 = (p * C_up + (1 - p) * C_down) / (1 + r)

print("Stock price if up:", S_up)
print("Stock price if down:", S_down)
print("Risk-neutral probability:", p)
print("Call option price today:", C0)
```

```python
import numpy as np

def binomial_option_price(S0, K, u, d, r, N):

    # risk-neutral probability
    p = ((1 + r) - d) / (u - d)

    # Step 1: stock prices at maturity
    stock_prices = np.zeros(N + 1)

    for j in range(N + 1):
        stock_prices[j] = S0 * (u**j) * (d**(N-j))

    # Step 2: option payoffs at maturity (European Call)
    option_values = np.maximum(stock_prices - K, 0)

    # Step 3: backward induction
    for i in range(N-1, -1, -1):
        for j in range(i+1):
            option_values[j] = (p * option_values[j+1] + (1-p) * option_values[j]) / (1+r)

    return option_values[0]


# Example parameters
S0 = 100
K = 100
u = 1.2
d = 0.8
r = 0.05
N = 7

price = binomial_option_price(S0, K, u, d, r, N)

print("Option price with 7 periods:", price)
```
