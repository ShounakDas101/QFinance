##Finite Probability Spaces


```python
import itertools

def generate_sample_space(n):
    return list(itertools.product(['H', 'T'], repeat=n))


def generate_atoms(omega, k):
    partition = {}

    for outcome in omega:
        key = outcome[:k]
        partition.setdefault(key, []).append(outcome)

    return list(partition.values())


def display_results(n):
    omega = generate_sample_space(n)

    print(f"\nSample Space Omega (n={n}):")
    for outcome in omega:
        print(outcome)

    # Show filtration
    for k in range(1, n + 1):
        print(f"\nF_{k} (atoms / partition based on first {k} tosses):")
        atoms = generate_atoms(omega, k)

        for i, atom in enumerate(atoms, 1):
            print(f"Atom {i}: {atom}")


# Run example
display_results(4)
```
