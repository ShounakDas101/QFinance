##Finite Probability Spaces


```python
import itertools

def generate_sample_space(n):
    return list(itertools.product(['H', 'T'], repeat=n))


def generate_atoms(omega, k):
    partition = {}

    for outcome in omega:
        key = outcome[:k]
        partition.setdefault(key, []).append(outcome)

    return list(partition.values())


def display_results(n):
    omega = generate_sample_space(n)

    print(f"\nSample Space Omega (n={n}):")
    for outcome in omega:
        print(outcome)

    # Show filtration
    for k in range(1, n + 1):
        print(f"\nF_{k} (atoms / partition based on first {k} tosses):")
        atoms = generate_atoms(omega, k)

        for i, atom in enumerate(atoms, 1):
            print(f"Atom {i}: {atom}")


# Run example
display_results(4)
```

# Conditional Probability

```python
import itertools

# Sample space
def generate_sample_space(n):
    return list(itertools.product(['H', 'T'], repeat=n))

# Partition (atoms of F_k)
def generate_atoms(omega, k):
    partition = {}
    for outcome in omega:
        key = outcome[:k]
        partition.setdefault(key, []).append(outcome)
    return list(partition.values())

# Define X_l (example: l-th coin as +1/-1)
def X_l(outcome, l):
    return 1 if outcome[l-1] == 'H' else -1

# Compute E[X_l | F_k]
def conditional_expectation(omega, k, l):
    atoms = generate_atoms(omega, k)

    result = {}

    for atom in atoms:
        # average X_l over the atom
        avg = sum(X_l(w, l) for w in atom) / len(atom)

        # assign same value to all outcomes in this atom
        for w in atom:
            result[w] = avg

    return result

# Display nicely
def display_conditional_expectation(n, k, l):
    omega = generate_sample_space(n)
    cond_exp = conditional_expectation(omega, k, l)

    print(f"\nE[X_{l} | F_{k}]:")

    for w in omega:
        print(f"{w} -> {cond_exp[w]}")

# Example
display_conditional_expectation(n=3, k=2, l=3)
```

# To verify Martingale Property

```python
import itertools

# Generate sample space
def generate_sample_space(n):
    return list(itertools.product(['H', 'T'], repeat=n))

# Generate atoms of F_k
def generate_atoms(omega, k):
    partition = {}
    for outcome in omega:
        key = outcome[:k]
        partition.setdefault(key, []).append(outcome)
    return list(partition.values())

# Define X_i
def X_i(outcome, i):
    return 1 if outcome[i-1] == 'H' else -1

# Define S_n (random walk)
def S_n(outcome, n):
    return sum(X_i(outcome, i) for i in range(1, n+1))

# Compute conditional expectation E[S_l | F_k]
def conditional_expectation_S(omega, k, l):
    atoms = generate_atoms(omega, k)
    result = {}

    for atom in atoms:
        avg = sum(S_n(w, l) for w in atom) / len(atom)
        for w in atom:
            result[w] = avg

    return result

# Verify martingale property
def verify_martingale(n, k, l):
    omega = generate_sample_space(n)

    cond_exp = conditional_expectation_S(omega, k, l)

    print(f"\nVerifying E[S_{l} | F_{k}] = S_{k}:\n")

    all_good = True

    for w in omega:
        lhs = cond_exp[w]
        rhs = S_n(w, k)

        print(f"{w} -> E[S_{l}|F_{k}] = {lhs}, S_{k} = {rhs}")

        if abs(lhs - rhs) > 1e-9:
            all_good = False

    if all_good:
        print("\n✅ Martingale property VERIFIED")
    else:
        print("\n❌ Martingale property FAILED")

# Example
verify_martingale(n=3, k=1, l=3)
```
